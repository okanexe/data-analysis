```python
import pandas as pd 
movies=pd.read_csv("movie_metadata.csv")
#movies.head()

df=movies[["movie_title","genres","plot_keywords","imdb_score"]]

#df1=df[df.isna().any(axis=1)]
#df1.sort_values(by=['imdb_score'],ascending=False)

df2=df[["movie_title","plot_keywords"]]
#df2.head()
keywords_list=[]
df2 = df2.dropna(how='any')
for row in df2["plot_keywords"]:
    keywords=row.split("|")
    for key in keywords:
        keywords_list.append(key)
len(keywords_list)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

count.update(Counter(combinations(keywords_list, 1)))

for key,value in count.most_common(10):
    print(key, value)
```

```python
df2=df[["movie_title","plot_keywords"]]
#df2.head()
keywords_list=[]
kelimeler=[]
liste=[]
df2 = df2.dropna(how='any')
for row in df2["plot_keywords"]:
    keywords=row.split("|")
    for key in keywords:
        keywords_list.append(key)
for k in keywords_list:
    keys=k.split(" ")
    for ke in keys:
        kelimeler.append(ke)
for k in kelimeler:
    if (len(k)>=3):
        liste.append(k)
```

```python

count = Counter()

count.update(Counter(combinations(liste, 1)))

for key,value in count.most_common(20):
    print(key, value)
```

```python
import pandas as pd 
from itertools import combinations
from collections import Counter


movies=pd.read_csv("movies_metadata.csv")
#movies["overview"].head(1)
keywords=[]
m_list=movies[["overview","title","vote_average"]].head(10000)
m_list=m_list.dropna(how='any')
for key in m_list["overview"]: 
    list=key.split(" ")
    for l in list:
        if(len(l)>=4):
            keywords.append(l)


count = Counter()

count.update(Counter(combinations(keywords, 1)))

for key,value in count.most_common(20):
    print(key, value)
```
