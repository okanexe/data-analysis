```python
import pandas as pd 
movies=pd.read_csv("movies_metadata.csv")

analysis = movies[["genres","imdb_id","overview","title"]]
#for row in analysis["genres"]:
    #print(row)
analysis.tail(20)
```

# fixed Imdb Link column as imdb_id because with imdb_id we can define the movies in every movie_dataset.csv . imdb_link is common attribute in every movies dataset .

```python
imdb_ids = pd.read_csv("MovieGenre.csv",encoding='latin-1')
ids_list=imdb_ids[["Imdb Link","Genre","Title"]]
ids=[]
def listToString(s):  
    str1 = "" 
    return (str1.join(s))

for row in ids_list["Imdb Link"]:
    liste=row.split("/")
    l=list(liste[4])
    l.insert(2,"0")
    ids.append(listToString(l))
    
ids_list["Imdb Link"]=ids
ids_list.tail(20)
```

# here we comparasion imdb_id and adjust Genre column as correct and what we want . 

```python
import numpy as np 

getAnalysis=analysis.merge(ids_list,how='inner',left_on='imdb_id',right_on='Imdb Link')
getAnalysis=getAnalysis[["imdb_id","overview","title","Genre","Title"]]
getAnalysis.tail(20)
```

```python
#getAnalysis['Genre'] = getAnalysis['Genre'].apply(lambda x: x.split("|"))
getAnalysis=getAnalysis.dropna(how='any')

getAnalysis['Genres'] = getAnalysis['Genre'].apply(lambda x: listToString(x).split("|"))
getAnalysis
```

## Find all genres in our dataframe 

```python
genre=[]
for row in getAnalysis["Genre"]:
    for gen in row:
        genre.append(gen)
genre=set(genre)
genre

```

# there is count of the genre how many times dupliceted 

```python
from itertools import combinations
from collections import Counter

count = Counter()

count.update(Counter(combinations(genre, 1)))

for key,value in count.most_common(10):
    print(key, value)
```

# There is Action movie overview analysis with how many words in overview ...

```python
ActionList=[]
for a,b in zip(getAnalysis["Genre"],getAnalysis["overview"]):
    try:
        if(a.index('Action')>=0):
            ActionList.append(b)
    except:
        pass
actionWords=[]
for w in ActionList:
    words=w.split(" ")
    for wrd in words:
        if(len(wrd)>3):
            actionWords.append(wrd)
englishGrammar=["with","that","from","their","they","into","when","When","must","this","have","after","only","find","will",
                "which","them","back","been","about","After","where","world","gets","film"]
for i in englishGrammar:
    for j in actionWords:
        if(i==j):
            actionWords.remove(j)
    
count = Counter()

count.update(Counter(combinations(actionWords, 1)))

for key,value in count.most_common(20):
        print(key,value)
```

# There is Comedy movie overview analysis with how many words in overview ...

```python
comedyList=[]
for a,b in zip(getAnalysis["Genre"],getAnalysis["overview"]):
    try:
        if(a.index('Comedy')>=0):
            comedyList.append(b)
    except:
        pass
comedyWords=[]
for w in comedyList:
    words=w.split(" ")
    for wrd in words:
        if(len(wrd)>3):
            comedyWords.append(wrd)
englishGrammar=["with","that","from","their","they","into","when","When","must","this","have","after","only","find","will",
                "which","them","back","been","about","After","where","world","gets","film","while","being","more","make"]
for i in englishGrammar:
    for j in comedyWords:
        if(i==j):
            comedyWords.remove(j)
    
count = Counter()

count.update(Counter(combinations(comedyWords, 1)))

for key,value in count.most_common(20):
        print(key,value)
```
